```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
import re

def to_snake(col):
    col = col.lower()                        # lowercase
    col = re.sub(r'[^a-z0-9]+', '_', col)    # replace spaces, brackets, etc.
    col = re.sub(r'_+', '_', col)            # remove multiple underscores
    col = col.strip('_')                     # remove leading/trailing _
    return col

df.columns = [to_snake(c) for c in df.columns]
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_usd':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create a column purchase_frequency_days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
